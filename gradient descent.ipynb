```python
currentx = 2
rate = 0.01
precision = 0.000001
delta_x = 1
max_iterations = 10000
iterations = 0
x = []
y = []
```

```python
while delta_x>precision and iterations<max_iterations:
    previous_x = current_x
    current_x = previous_x - rate * 2 *(previous_x+3)
    y.append((current_x+3)**2)
    x.append(current_x)
    delta_x = abs(previous_x - current_x)
    iterations +=1
print(current_x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```
